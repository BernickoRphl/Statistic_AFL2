```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Menggunakan path lokal ke file Excel
file_path = "AFL_2.xlsx"

# Membaca file Excel dari path lokal
df = pd.read_excel(file_path)

# Mengubah atribut kolom
df.columns = ['Name', 'Artist', 'Year', 'Genre', 'Rating', 'Popularity', 'Royalty', 'Streams']

# Scatter plot: Rating vs Popularity
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Rating', y='Popularity')
plt.title('Rating vs Popularity')
plt.xlabel('Rating')
plt.ylabel('Popularity')
plt.grid(True)
plt.show()

# Box plot: Gaji berdasarkan Genre
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Genre', y='Royalty')
plt.title('Gaji Berdasarkan Genre')
plt.xlabel('Genre')
plt.ylabel('Gaji')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Histogram: Distribusi tahun perilisan lagu
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Year', bins=range(2015, 2023), discrete=True, kde=False)
plt.title('Distribusi Tahun Perilisan Lagu')
plt.xlabel('Tahun')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
# Plot diagram lingkaran persentase genre
plt.figure(figsize=(8, 8))
df['Genre'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Persentase Genre')
plt.ylabel('')
plt.show()
```

```python
# Plot histogram umur artist
plt.figure(figsize=(8, 6))
plt.hist(df['Year'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribusi Tahun Rilis Lagu')
plt.xlabel('Tahun Rilis')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```

```python
#Hipotesis 1: Perbedaan Rata-rata Rating antara Lagu Pop dan Lagu Electronic (Two-sample)
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x='Genre', y='Rating', data=df[df['Genre'].isin(['Pop', 'Electronic'])])
plt.title('Perbedaan Rata-rata Rating antara Lagu Pop dan Lagu Electronic')
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.show()
```

```python
#Hipotesis 2: Rata-rata Rating Lagu Pop pada Tahun 2020 (One-sample)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Year', y='Rating', data=df[df['Genre'] == 'Pop'])
plt.title('Rata-rata Rating Lagu Pop pada Tahun 2020')
plt.xlabel('Tahun')
plt.ylabel('Rating')
plt.show()
```

```python
#Hipotesis 3: Perbedaan Rata-rata Popularitas antara Lagu Indie dan Lagu Folk (Two-sample)
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Genre', y='Popularity', data=df[df['Genre'].isin(['Indie', 'Folk'])])
plt.title('Perbedaan Rata-rata Popularitas antara Lagu Indie dan Lagu Folk')
plt.xlabel('Genre')
plt.ylabel('Popularity')
plt.show()
```
